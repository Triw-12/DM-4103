```python
import networkx as nx
from sklearn.metrics import adjusted_mutual_info_score, normalized_mutual_info_score

def detect_communities_louvain(graph):
    communities = nx.community.louvain_communities(graph)

    partition = {}
    for community_id, community_nodes in enumerate(communities):
        for node in community_nodes:
            partition[node] = community_id

    # Préparation des labels
    # On crée les listes ordonnées pour la comparaison
    node_list = list(graph.nodes())
    algo_labels = [partition[node] for node in node_list]
    # On récupère les valeurs des attributs depuis les attributs des nœuds
    dorm_labels = [graph.nodes[node]['dorm'] for node in node_list]
    major_labels = [graph.nodes[node]['major_index'] for node in node_list]

    # Évaluation
    ami_score_dorm = adjusted_mutual_info_score(dorm_labels, algo_labels)
    nmi_score_dorm = normalized_mutual_info_score(dorm_labels, algo_labels)
    ami_score_major = adjusted_mutual_info_score(major_labels, algo_labels)
    nmi_score_major = normalized_mutual_info_score(major_labels, algo_labels)

    print(f"Nombre de communautés détectées : {len(communities)}")
    print(f"Score AMI pour dorm : {ami_score_dorm}")
    print(f"Score NMI pour dorm : {nmi_score_dorm}")
    print(f"Score AMI pour major : {ami_score_major}")
    print(f"Score NMI pour major : {nmi_score_major}")

print("Résultats pour le graphe MIT8 :")
detect_communities_louvain(nx.read_gml("data/MIT8.gml"))
print("\nRésultats pour le graphe Caltech36 :")
detect_communities_louvain(nx.read_gml("data/Caltech36.gml"))
print("\nRésultats pour le graphe Johns Hopkins55 :")
detect_communities_louvain(nx.read_gml("data/Johns Hopkins55.gml"))
```
